Day 2: Handle Categorical Data - Apply Label Encoding and One-Hot Encoding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load Dataset
df = pd.read_csv("C:/Terraform/TRUPTI/The  developer arena/week10-Customer-churn-prediction/customer_churn.csv")

print("Original Dataset")
print(df.head())

# Create LabelEncoder
le = LabelEncoder()

# Label Encoding
df["Contract"] = le.fit_transform(df["Contract"])
df["PaymentMethod"] = le.fit_transform(df["PaymentMethod"])
df["PaperlessBilling"] = le.fit_transform(df["PaperlessBilling"])

print("\nAfter Label Encoding")
print(df.head())

# One-Hot Encoding
df = pd.get_dummies(df, columns=["Contract"], dtype=int)

print("\nAfter One-Hot Encoding")
print(df.head())
print(df)
# Display Updated Columns
print("\nUpdated Columns")
print(df.columns)
```

Day 4: Outlier Detection & Handling - Identify outliers using IQR and Z-score methods

```python
# Import Libraries
import pandas as pd
import numpy as np
from scipy.stats import zscore

# Load Dataset
df = pd.read_csv("C:/Terraform/TRUPTI/The  developer arena/week10-Customer-churn-prediction/customer_churn.csv")

# IQR Method
Q1 = df["MonthlyCharges"].quantile(0.25)
Q3 = df["MonthlyCharges"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df["MonthlyCharges"] < lower) | (df["MonthlyCharges"] > upper)]
print("Outliers using IQR Method")
print(outliers)


# Z-Score Method
df["ZScore"] = zscore(df["MonthlyCharges"])

z_outliers = df[df["ZScore"] > 3]

print("\nOutliers using Z-Score Method")
print(z_outliers)
```

Day 5: Feature Engineering - Create new features like Customer Lifetime Value, Payment Efficiency

```python
# Import Library
import pandas as pd

# Load Dataset
df = pd.read_csv("C:/Terraform/TRUPTI/The  developer arena/week10-Customer-churn-prediction/customer_churn.csv")

# Create New Features

# 1. Charges Per Month
df["ChargesPerMonth"] = df["TotalCharges"] / (df["Tenure"] + 1)

# 2. Customer Lifetime Value
df["CustomerLifetimeValue"] = df["MonthlyCharges"] * df["Tenure"]

# 3. Senior Citizen Score
df["SeniorCitizenScore"] = df["SeniorCitizen"] * 10

# 4. Paperless Billing Score
df["PaperlessScore"] = df["PaperlessBilling"].map({"Yes":1,"No":0})

# 5. Long Term Customer
df["LongTermCustomer"] = (df["Tenure"] > 24).astype(int)

# Display Dataset
print(df.head())
```

Day 7: Build Pipeline - Create complete preprocessing pipeline, test end-to-end

```python
# Import Libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load Dataset
df = pd.read_csv("C:/Terraform/TRUPTI/The  developer arena/week10-Customer-churn-prediction/customer_churn.csv")

le = LabelEncoder()

df["Contract"] = le.fit_transform(df["Contract"])
df["PaymentMethod"] = le.fit_transform(df["PaymentMethod"])
df["PaperlessBilling"] = le.fit_transform(df["PaperlessBilling"])
df["Churn"] = le.fit_transform(df["Churn"])

# Feature Scaling

scaler = StandardScaler()

df[["Tenure", "MonthlyCharges", "TotalCharges"]] = scaler.fit_transform(
    df[["Tenure", "MonthlyCharges", "TotalCharges"]]
)

# Display Final Dataset
print("Preprocessing Pipeline Completed Successfully")
print(df.head())
```
